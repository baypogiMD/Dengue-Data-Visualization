# Import Libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Dataset

```
Philippines = pd.read_csv('dengue_climate.csv')
```

## National Capital Region

```
NCR = Philippines[(Philippines['Region'] == 'NCR')]
NCR_hum = NCR.iloc[:, 4]
NCR_rain = NCR.iloc[:, 5]
NCR_temp = NCR.iloc[:, 6]
NCR_deng = NCR.iloc[:, 7]
```

```
fig, axs = plt.subplots(1,3, sharey=True, figsize=(12,3), dpi=150)
fig.suptitle('National Capital Region')

axs[0].plot(NCR_hum, NCR_deng, 's', color='green')
m, b = np.polyfit(NCR_hum, NCR_deng, 1)
axs[0].plot(NCR_hum, m*NCR_hum + b, color='orange')
axs[0].set_xlabel('Humidity (%)')
axs[0].set_ylabel('Dengue Incidence')

axs[1].plot(NCR_rain, NCR_deng, 'v', color='darkblue')
m, b = np.polyfit(NCR_rain, NCR_deng, 1)
axs[1].plot(NCR_rain, m*NCR_rain + b, color='orange')
axs[1].set_xlabel('Rainfall (mm)')

axs[2].plot(NCR_temp, NCR_deng, 'o', color='red')
m, b = np.polyfit(NCR_temp, NCR_deng, 1)
axs[2].plot(NCR_temp, m*NCR_temp + b, color='orange')
axs[2].set_xlabel('Temperature (C)')

#plt.savefig('ncr.jpg', dpi=150)
plt.show()
```
